```python
import numpy as np
import matplotlib.pyplot as plt
import cmath

def f(x) :
    return x ** 4 - 11 * x + 8
    
num_iter = 20
p0 = 1
p1 = 2
q0 = f(p0)
q1 = f(p1)
assert(q0 * q1 < 0)
for it in range(num_iter) :
    p2 = p1 - q1 * (p1 - p0) / (q1 - q0)
    q2 = f(p2)
    print(it + 1, p2, q2)
    if(q2 * q1 < 0) :
        p0 = p1
        q0 = q1
        p1 = p2
        q1 = q2
    else :
        p1 = p2
        q1 = q2
```
