```python
# Newton's method for non - linear system of equations
# X(k+1) = X(k) - inv(J(k)) * F(X(k))
# J is jacobian matrix

# 2D
# f(x, y) = 0
# g(x, y) = 0
# solve for x and y

def f(X) :
    x = X[0]
    y = X[1]
    return x * x + x * y + y * y - 7
    
def g(X) :
    x = X[0]
    y = X[1]
    return x ** 3 + y ** 3 - 9

def fx(X) :
    x = X[0]
    y = X[1]
    return 2 * x + y
    
def fy(X) :
    x = X[0]
    y = X[1]
    return x + 2 * y
    
def gx(X) :
    x = X[0]
    y = X[1]
    return 3 * x * x

def gy(X) :
    x = X[0]
    y = X[1]
    return 3 * y * y
    
# initial approximation
x0 = 1.5
y0 = 0.5
X = [1.5, 0.5]
F = [f(X), g(X)]

num_iter = 20
for it in range(num_iter) :
    jacob = np.array([[fx(X), fy(X)], [gx(X), gy(X)]])
    inv_jacob = np.linalg.inv(jacob)
    nX = X - np.dot(inv_jacob, F)
    X = nX
    F = [f(X), g(X)]
    print(it + 1, X, F)
```
